**Aim : Evaluate the effectiveness of cosine similarity and Jaccard similarity measures in
ranking documents within an Information Retrieval system**

```python
import math
from collections import Counter
```

```python
docs = {
    "D1": "information retrival system vector",
    "D2": "information retrival similarity measure",
    "D3": "vector space model retrival"
}
```

```python
query = "information retrival vector"
```

```python
# Tokenize
def tokenize(text):
    return text.split()
```

```python
# Build vocabulary
vocab = set()
for d in docs.values():
    vocab.update(tokenize(d))
vocab.update(tokenize(query))
vocab = list(vocab)
```

```python
def vector(text):
    words = Counter(tokenize(text))
    return [words[w] for w in vocab]
```

```python
# Cosine similarity
def cosine(v1, v2):
    dot = sum(a*b for a, b in zip(v1, v2))
    mag1 = math.sqrt(sum(a*a for a in v1))
    mag2 = math.sqrt(sum(b*b for b in v2))
    return dot/(mag1*mag2) if mag1 and mag2 else 0
```

```python
# Jaccard similarity
def jaccard(t1, t2):
    s1, s2 = set(tokenize(t1)), set(tokenize(t2))
    return len(s1 & s2) / len(s1 | s2)
```

```python
# Compute and rank
print("Cosine Similarity Ranking:")
cos_scores = {k: cosine(vector(v), vector(query)) for k, v in docs.items()}
for d in sorted(cos_scores, key=cos_scores.get, reverse=True):
    print(d, round(cos_scores[d], 4))

print("\nJaccard Similarity Ranking:")
jac_scores = {k: jaccard(v, query) for k, v in docs.items()}
for d in sorted(jac_scores, key=jac_scores.get, reverse=True):
    print(d, round(jac_scores[d], 4))
```
