```python
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Load dataset
data = pd.read_csv('/content/drive/MyDrive/ML Dataset/insurance[1].csv')

# Convert smoker column to numeric
data['smoker'] = data['smoker'].map({'yes':1, 'no':0})

# Actual relevant
actual = data['smoker']

# Retrieved documents rule
retrieved = (data['charges'] > data['charges'].mean()).astype(int)

print("Confusion Matrix:\n", confusion_matrix(actual, retrieved))
print("Accuracy:", accuracy_score(actual, retrieved))
print("Precision:", precision_score(actual, retrieved))
print("Recall:", recall_score(actual, retrieved))
print("F1 Score:", f1_score(actual, retrieved))
```
