**To implement the TF–IDF weighting technique for a given document corpus and generate a TF–IDF matrix for effective document representation.**

```python
import math
import pandas as pd
```

```python
# Sample documents
documents = [
    "machine learning is fun",
    "machine learning is powerful",
    "python is used for machine learning"
]
```

```python
# Tokenize documents
tokenized_docs = [doc.split() for doc in documents]
```

```python
# Vocabulary
vocab = sorted(set(word for doc in tokenized_docs for word in doc))
```

```python
#Name : Ashwini Khushal Barapatre
# ---------------- TF ----------------
tf = []
for doc in tokenized_docs:
    tf_doc = {}
    total_words = len(doc)

    for word in vocab:
        tf_doc[word] = doc.count(word) / total_words

    tf.append(tf_doc)

tf_df = pd.DataFrame(tf)

print("1) TERM FREQUENCY (TF)\n")
print(tf_df)
print("\n")
```

```python
# Name : Ashwini Khushal Barapatre
# ---------------- IDF ----------------
idf = {}
N = len(tokenized_docs)

for word in vocab:
    doc_count = sum(1 for doc in tokenized_docs if word in doc)
    idf[word] = math.log(N / doc_count)

idf_df = pd.DataFrame(list(idf.items()), columns=["Word", "IDF"])

print("2) INVERSE DOCUMENT FREQUENCY (IDF)\n")
print(idf_df)
print("\n")
```

```python
# Name : Ashwini Khushal Barapatre
# ---------------- TF-IDF ----------------
tfidf = []

for doc_tf in tf:
    tfidf_doc = {}
    for word in vocab:
        tfidf_doc[word] = doc_tf[word] * idf[word]

    tfidf.append(tfidf_doc)

tfidf_df = pd.DataFrame(tfidf)

print("3) TF-IDF MATRIX\n")
print(tfidf_df)
```
