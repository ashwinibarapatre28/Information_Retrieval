```python
import nltk
import re
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
text = """
Information Retrieval is the process of obtaining relevant information from large collections of data.
Text preprocessing plays a major role in improving search results by cleaning and normalizing data.
"""
print("Original Text:\n", text)
```

```python
sentences = sent_tokenize(text)
print("Sentence Tokens:\n", sentences)
```

```python
words = word_tokenize(text)
print("Word Tokens:\n", words)
```

```python
clean_words = [re.sub(r'[^a-zA-Z]', '', word).lower() for word in words]
clean_words = [word for word in clean_words if word != '']
print("Cleaned Words:\n", clean_words)
```

```python
stop_words = set(stopwords.words('english'))

filtered_words = [word for word in clean_words if word not in stop_words]
print("After Stop-word Removal:\n", filtered_words)
```

```python
ps = PorterStemmer()

stemmed_words = [ps.stem(word) for word in filtered_words]
print("After Stemming:\n", stemmed_words)
```

```python
print("----- Comparison -----")
print("Original Words:\n", clean_words)
print("\nAfter Stop-word Removal:\n", filtered_words)
print("\nAfter Stemming:\n", stemmed_words)
```
